```python
import yfinance as yf
import matplotlib.pyplot as plt
```

<h1>Get the data for the stock Apple, Google</h1>

```python
dataAAPL = yf.download('AAPL', '2023-01-01', '2023-10-31')
dataGOOG = yf.download('GOOG', '2023-01-01', '2023-10-31')
```

<h1>Show data Apple, Google</h1>

```python
dataGOOG.head(10)
```

```python
dataAAPL.describe()
```

```python
dataGOOG.describe()
```

<h1>Visualize Open and Close Apple</h1>

```python
date =  dataAAPL.index
plt.plot(date, dataAAPL.Close, label = "Close")
plt.plot(date, dataAAPL.Open, label = "Open")
plt.legend(title = "Apple Inc", loc = 'upper left')
fig = plt.gcf()
fig.set_size_inches(20,10)
plt.show
```

<h1>Visualize Open and Close Google</h1>


```python
date =  dataAAPL.index
plt.plot(date, dataGOOG.Close, label = "Close")
plt.plot(date, dataGOOG.Open, label = "Open")
plt.legend(title = "Google Inc", loc = 'upper left')
fig = plt.gcf()
fig.set_size_inches(20,10)
plt.show
```

<h1>Visualze Open Apple, Google</h1>

```python
date =  dataAAPL.index
plt.plot(date, dataGOOG.Open, label = "Google")
plt.plot(date, dataAAPL.Open, label = "Apple")
plt.legend(title = "Company")
fig = plt.gcf()
fig.set_size_inches(20,10)
plt.show
```

<h1>Visualze volume Apple, Google</h1>

```python
date =  dataAAPL.index
plt.plot(date, dataAAPL.Volume, label = "Apple")
plt.plot(date, dataGOOG.Volume, label = "Google")
plt.legend(title = "Volume", loc = 'upper left')
fig = plt.gcf()
fig.set_size_inches(20,10)
plt.show
```
